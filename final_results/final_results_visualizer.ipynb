```python
import matplotlib.pyplot as plt
```

```python
def parse_data(filename):
    """
    Parses data from a text file into separate lists for epochs, loss, val, and test.

    Args:
        filename: The name of the text file containing the data.

    Returns:
        A dictionary containing four lists: epochs, loss, val, and test.
    """
    epochs = []
    loss = []
    val = []
    test = []

    with open(filename, 'r') as f:
        for line in f:
            # Split the line based on commas and colons
            parts = line.strip().split(":")
            # print(parts)
            if len(parts) != 5:
                # Skip lines with unexpected format
                continue

            # Extract epoch (convert to integer)
            epoch = int(parts[1].split(",")[0])
            epochs.append(epoch)

            # Extract values (convert to float)
            loss.append(float(parts[2].split(",")[0]))
            val.append(float(parts[3].split(",")[0]))
            test.append(float(parts[4].split(",")[0]))

    return {'epochs': epochs, 'loss': loss, 'val': val, 'test': test}

# Example usage
data = parse_data("final_results.txt")  # Replace with your actual filename
```

```python
# Data
epochs = data["epochs"]
loss = data["loss"]
val = data["val"]
test = data["test"]
```

```python
# Line plot
plt.figure(figsize=(10, 6))

plt.plot(epochs, loss, label='Loss')
plt.plot(epochs, val, label='Val')
plt.plot(epochs, test, label='Test')

# Labels and title
plt.xlabel('Epoch')
plt.ylabel('Value')
plt.title('Normal ZINC Performance')

# Legend
plt.legend()

# Grid
plt.grid(True)

# Show plot
plt.show()
```
